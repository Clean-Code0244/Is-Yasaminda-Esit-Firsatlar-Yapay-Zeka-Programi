# Görüldüğü üzere veri setinde eksik bir veri bulunmamaktadır.

```python
drop_columns = ['img_link', 'product_link', 'about_product', 'user_name']
df.drop(columns=drop_columns, inplace=True)
```

# Gereksiz sütunları sildim.

```python
df.shape
```

```python
df['actual_price'] = df['actual_price'].astype(str).replace('₹', '', regex=True).str.replace(',', '', regex=False).astype(float)
df['actual_price_in_try'] = df['actual_price'] * 0.42
```

# Burda hindistan rupisi biriminde olan paraları TL birimine çevirmek için önce stringe çevirip para birimlerini kaldırdım, sonrasında ise günümüzde 1 rupi 0,42 tl olduğu için sütundaki tüm değerleri 0,42 ile çarparak price_in_try adlı yeni bir sütun açtım.

```python
df['actual_price_in_try'].head(20)
```

```python
df['discounted_price'] = df['discounted_price'].astype(str).replace('₹', '', regex=True).str.replace(',', '', regex=False).astype(float)
df['discounted_price_in_try'] = df['discounted_price'] * 0.42
```

```python
df['discounted_price_in_try'].head(20)
```

```python
df['discount_percentage'] = df['discount_percentage'].replace('%', '', regex=True).astype(float)
```

```python
df['discount_percentage'].head(10)
```

Ortalama İndirim Oranı

```python
print("Ortalama İndirim Oranı: ", df['discount_percentage'].mean())
```

#  Kategori Bazında Satış Analizi

```python
category_counts = df['category'].value_counts()
print("En Popüler Kategoriler:")
top_10_categories = category_counts.head(10)
print(top_10_categories)
```

```python

top_10_categories.plot(kind='bar', color='skyblue')

plt.title("En Popüler Kategoriler")
plt.xlabel("Kategori")
plt.ylabel("Frekans")
plt.xticks(rotation=45, ha='right')  # X eksenindeki kategorileri düzgün görmek için
plt.show()
```

# En Yüksek Puanlı Ürünler

```python

df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# 4.5 ve üzeri puan alan ürünleri seçip, rating_count'a göre azalan sırayla sıralama
top_rated_products = df[df['rating'] >= 4.5].sort_values(by='rating_count', ascending=False)

# En popüler 10 ürün
print("En Çok Puan Alan Ürünler:")
print(top_rated_products[['product_name', 'rating', 'rating_count']].head(10))
```

```python

# İlk 10 ürünün verisini aldım
top_10_rated_products = top_rated_products[['product_name', 'rating', 'rating_count']].head(10)


labels = top_10_rated_products['product_name']
sizes = top_10_rated_products['rating_count']


plt.figure(figsize=(10, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)

plt.title("En Çok Puan Alan 10 Ürünün Oy Sayıları", fontsize=16)


plt.axis('equal')
plt.show()
```

# En Çok Yorum Alan Ürünler

```python
top_reviewed_products = df.sort_values(by='rating_count', ascending=False)
print("En Çok Yorum Alan Ürünler:")
print(top_reviewed_products[['product_name', 'rating_count']].head(10))
```

```python
# İlk 10 en çok yorum alan ürünü aldım
top_10_reviewed_products = top_reviewed_products[['product_name', 'rating_count']].head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_10_reviewed_products['product_name'], top_10_reviewed_products['rating_count'], color='lightgreen')

plt.title("En Çok Yorum Alan 10 Ürün", fontsize=16)
plt.xlabel("Yorum Sayısı", fontsize=12)
plt.ylabel("Ürün Adı", fontsize=12)

plt.show()
```

# En Yüksek 10 İndirim Yüzdesine Sahip Ürünler

```python


# En yüksek 10 indirim yüzdesine sahip ürünleri aldım
top_10_discounted_products = df[['product_name', 'discount_percentage']].sort_values(by='discount_percentage', ascending=False).head(10)

labels = top_10_discounted_products['product_name']
sizes = top_10_discounted_products['discount_percentage']


plt.figure(figsize=(10, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)

plt.title("En Yüksek 10 İndirim Yüzdesine Sahip Ürünler", fontsize=16)

plt.axis('equal')
plt.show()
```

#  En pahalı ve En ucuz 10 Ürün

```python
# En pahalı 10 ürün
top_10_expensive_products = df[['product_name', 'actual_price_in_try']].sort_values(by='actual_price_in_try', ascending=False).head(10)

# En ucuz 10 ürün
top_10_cheapest_products = df[['product_name', 'actual_price_in_try']].sort_values(by='actual_price_in_try', ascending=True).head(10)

# Sonuçları yazdır
print("En Pahalı 10 Ürün:")
print(top_10_expensive_products)

print("\nEn Ucuz 10 Ürün:")
print(top_10_cheapest_products)
```

```python

plt.figure(figsize=(10, 6))
sns.barplot(x='actual_price_in_try', y='product_name', data=top_10_expensive_products, palette='Reds')
plt.title("En Pahalı 10 Ürün", fontsize=16)
plt.xlabel("Fiyat (TRY)", fontsize=12)
plt.ylabel("Ürün Adı", fontsize=12)
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='actual_price_in_try', y='product_name', data=top_10_cheapest_products, palette='Greens')
plt.title("En Ucuz 10 Ürün", fontsize=16)
plt.xlabel("Fiyat (TRY)", fontsize=12)
plt.ylabel("Ürün Adı", fontsize=12)
plt.show()
```
